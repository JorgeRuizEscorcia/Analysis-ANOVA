```python
import numpy as np
import pandas as pd
from scipy.stats import f_oneway
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns

```

```python
df.drop(columns=["ID"], inplace=True)
```

```python
df.shape
```

```python
num_vars = df[['GPA_CE',	'GPA_LC',	'GPA_RC',	'GPA_CC', 'GPA_T']]
cat_var = df['KOLB']
```

```python
for var in num_vars:
    groups = []
    for category in set(cat_var):
        groups.append(num_vars[cat_var == category][var])
    statistic, pvalue = f_oneway(*groups)
    print("Numerical Variable:", var)
    print("F-statistic:", statistic)
    print("p-value:", pvalue)
    print("\n")
```

```python
data = []
data.extend([{var: value, 'Category': category} for value, category in zip(groups, set(cat_var))])
```
